```python
# In order to speed up future data conversions, we pass a function to be able to convert a list of columns to numeric
def columns_to_numeric(df:pd.DataFrame,cols:list)->pd.DataFrame:
    ''' 
    This columns takes a list of columns from a DataFrame and converts them into numeric.
    Input:
    
    df: dataframe we want to clean
    cols: list of columns we want to convert to numeric.

    Output:
    colums properly formatted to numeric.
    '''
    
    df2 = df.copy()

    # we interate over the list of columns we want to convert into numeric
    for col in cols:
        df2[col] = pd.to_numeric(df2[col], errors='coerce')

    return df2

    
```

```python
# We will create a function to extract the day as day of the week, month, quarter and week of the year. This should help us analyse the seasonality

def date_extract(df:pd.DataFrame,col:"string")->pd.DataFrame:
    '''
    This function takes a DataFrame and extracts the month, quarter and week of the year into new columns

    Input:

    df: The dataframe we want to apply the changes on
    col: The date column we want to extract the month, quarter, week of the year into new colums from

    Output:

    it returns the dataframe with the new 3 columns
    '''
    # We create a copy of our df and extract the corresponding values from the date colum inserted as an input:
    df2=df.copy()
    df2["month"] = df2[col].dt.month
    df2["quarter"] =  df2[col].dt.quarter
    df2["week_of_year"]= df2[col].dt.isocalendar().week

    return df2

```

```python
# We will split the "demographic" target_audience into 2 columns, provided a specific pattern is present. In this case we will look for digit-digit

import pandas as pd

def extract_pattern(df: pd.DataFrame, col: str, pattern: str, new_col: str) -> pd.DataFrame:
    '''
    This function extracts a pattern from a specified column into a new column.

    Input:
    - df: The DataFrame we want to apply the changes to.
    - col: The column containing the text to search for the pattern.
    - pattern: The regular expression pattern to extract.
    - new_col: The name of the new column to store the extracted patterns.

    Output:
    - Returns the DataFrame with the new column.
    '''
    df2 = df.copy()

    # We use the methode str. to extract the pattern into a new column
    df2[new_col] = df2[col].str.extract(pattern, expand=False)

    # If there are missing values in the original column, fill them in the new column
    df2[new_col].fillna("all ages", inplace=True)

    return df2
```

```python
# We now will define a function to clean the remaining target_audience column to only have men, women and all genders

def gender_cleaning(df:pd.DataFrame,col:"string")->pd.DataFrame:
    '''
    This function takes a dataframe as input and returns 3 values for gender, men, women and all genders

    Input: 
    df: dataframe to be modified
    col: name of the column we want to apply the changes on

    Output:

    Returns the dataframe with the 3 options listed above: men, women and all genders.

    '''
    df2=df.copy()
    # Using the lambda function from the data transformation notebook, we adapt the function to the current purpose
    df2 = df.copy()
    df2[col] = df2[col].apply(lambda val: "women" if "women" in str(val).lower() else ("men" if "men" in str(val).lower() else "all genders"))

    return df2
```

```python
def line_plot_variables(df:pd.DataFrame,col_x:"string",col_y:"string",col_hue:"hue")-> None:
    '''
    This function takes columns x,y and a hue value from a dataframe and plots a line bar plot over a time range

    Input:
    df: dataframe to be plotted
    col_x: name of the column to pass on axis x
    col_y: name of the column to pass on axis y
    col_hue: which column in the data frame should be used for colour encoding. 
    '''
    df2=df.copy()
    fig, ax = plt.subplots(figsize=(12,6))
    sns.lineplot(data=df2, x=col_x, y=col_y, hue=col_hue,errorbar=None, ax=ax)
    plt.tight_layout()
    plt.show()
    return None
```

```python
# this is just an example to test the previous function:
line_plot_variables(marketing_df,"month", "cpm", "channel_used")
```

```python
def line_plot_multiple_variables(df:pd.DataFrame,col_x:"string",col_y:"string",list_col_hue:list)-> None:
    '''
    This function takes columns x,y and and a list for the hue value from a dataframe and plots as many line plots as columns in the hue list.

    Input:
    df: dataframe to be plotted
    col_x: name of the column to pass on axis x
    col_y: name of the column to pass on axis y
    list_col_hue: which columns in the data frame should be used for colour encoding. line_plot_variables.
    '''
    # We create a counter to be able to draw as many frames for plots as elements contained within the list_col_hue
    counter = 0
    # We pass the number of subplots applying a len into the list_col_hue
    fig, ax = plt.subplots(len(list_col_hue), figsize=(10, 5 * len(list_col_hue)))
    df2 = df.copy()
    # I create a loop for each col in the dataframe:
    for col in list_col_hue:
            sns.lineplot(data=df2, x=col_x, y=col_y, hue=col,errorbar=None, ax = ax [counter])
            # This will set the title to the visual
            ax[counter].set_title(col)
            counter +=1
    plt.tight_layout()
    plt.show()
    return None
```

### Conclusions:

##### Data interpretation:

- each row represents an individual campaign, each with an individual campaign ID. The campaigns have:
- - Company = the advertiser, our agency's client
  - Campaign type = the advertiser's own campaign topic
  - Target audience = the demographic targeting applied
  - Duration = the campaign's overall flight time
  - Channel used = where we served the ads
  - Conversion rate = % of conversions over the served impressions
  - Acquisition Cost = the advertiser's cost for each campaign
  - ROI = the advertiser's calculated return on investments, in $, for each dollar spent - calculated on advertiser's own purchase values not reported in this dataset
  - Location = where the campaign was served
  - Language = the creative's language
  - Clicks, Impressions = total ad impressions and clicks of each campaign
  - Engagement score = a rank 1-10 of user engagement, calculated on metrics not reported in this dataset
  - Customer segment = the interest based audience at which the campaign was targeted
  - Date = date in which the campaign was served

##### Data manipulation:
    
- We found that the Acquisition Cost column had likely a typo, and that the "," is to interpret as a decimal separator instead of thousands separator, so we replaced the data accordingly. This way, each row indicates the individual campaign spend. The sum of the total campaign spends returns our gross yearly revenue.
- We have split the target audience column into separate columns for gender and age range.
- We have extended the date information adding columns for month, quarter and week of the year.
- We have computed CTR and CPM from the provided data to further analyze the traffic cost and effectiveness.
